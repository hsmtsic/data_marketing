```python
import pandas as pd
import numpy as np
# 多重代入法 (IterativeImputer) のためにインポート
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# -----------------------------------------------------------------
# 1. データの読み込み
# -----------------------------------------------------------------
file_id = "1eui6vNdxLSCY8W_C1YSh5V9vBEuVhG-3"
download_url = f"https://drive.google.com/uc?id={file_id}"

try:
    # データ読み込み
    df = pd.read_pickle(download_url)
    print(f"--- 読み込み成功: 元のデータサイズ {df.shape} ---")

    # -----------------------------------------------------------------
    # 2. データクレンジングの実行
    # -----------------------------------------------------------------

    # 処理1: 異常値（マイナス・ゼロ）の除去
    # 根拠: アラート画像の「has zeros (0.5% - 2.0%)」
    #      および分析目的 (売上計算のため Quantity と UnitPrice は 0 より大きい必要がある)
    df_cleaned = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)].copy()
    print(f"--- 処理1 (ゼロ/マイナス除去)後: データサイズ {df_cleaned.shape} ---")

    # 処理2: 欠損値の除去 (CustomerID)
    # 根拠: アラート画像の「CustomerID has 135080 missing values (24.9%)」
    #      [cite_start]および資料の指示 (RFM分析 [cite: 70-75] にCustomerIDが必須のため)
    #      [cite_start]資料 [cite: 28] の「行と列の削除で対応」を適用
    df_cleaned.dropna(subset=['CustomerID'], inplace=True)
    print(f"--- 処理2 (CustomerID欠損除去)後: データサイズ {df_cleaned.shape} ---")

    # 処理3: 欠損値の補完 (Description)
    # 根拠: アラート画像の「Description has 1454 missing values (0.3%)」
    #      [cite_start]および資料の指示 (アソシエーション分析 [cite: 81] のため)
    #      [cite_start]資料 [cite: 30] の「補完」を適用 (カテゴリ変数のため最頻値や固定値で補完)
    if 'Description' in df_cleaned.columns:
        df_cleaned['Description'].fillna('Unknown', inplace=True)
        print(f"--- 処理3 (Description補完)完了 ---")

    # 処理4: 外れ値の除去 (IQR法)
    # 根拠: アラート画像の「Quantity is highly skewed」「UnitPrice is highly skewed」
    #      歪度が非常に高いため、平均値/標準偏差法は不適
    #      [cite_start]資料 [cite: 45] の「中央値と四分位範囲を使う方法」を適用
    cols_for_iqr = ['Quantity', 'UnitPrice']

    # (参考) IQR適用前の統計量
    # print("\n--- IQR適用前の統計量 (Q, UP) ---")
    # print(df_cleaned[cols_for_iqr].describe())

    for col in cols_for_iqr:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # 範囲内のデータのみを残す
        df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]

    print(f"--- 処理4 (IQR外れ値除去)後: データサイズ {df_cleaned.shape} ---")

    # 処理5: 多重代入法 (必要な場合のみ)
    # [cite_start]資料 [cite: 35] に記載の「多重代入法」
    # (注: ここまでの処理で主要な数値列の欠損は除去されている可能性が高い)

    numeric_cols = df_cleaned.select_dtypes(include=np.number).columns
    missing_in_numeric = df_cleaned[numeric_cols].isnull().sum().sum()

    if missing_in_numeric > 0:
        print(f"--- 処理5 (多重代入法) を {numeric_cols} に実行します ---")
        imputer = IterativeImputer(max_iter=10, random_state=0)

        # 補完を実行
        df_cleaned[numeric_cols] = imputer.fit_transform(df_cleaned[numeric_cols])
    else:
        print("--- 処理5 (多重代入法): 数値列に欠損値がないためスキップしました ---")


    # -----------------------------------------------------------------
    # [cite_start]3. 最終的な基礎統計量の出力 (資料 [cite: 66] の指示)
    # -----------------------------------------------------------------
    print("\n\n--- クレンジング完了後のデータ ---")

    print("\n--- 最終的なデータ情報 (df.info()) ---")
    df_cleaned.info()

    print("\n--- 最終的な基本統計量 (df.describe()) ---")
    print(df_cleaned.describe())

    print("\nクレンジングが完了しました。")

    # クレンジング後のDataFrameをCSVとして保存し、ユーザーが確認できるようにする
    df_cleaned.to_csv("cleansed_online_retail.csv", index=False)
    print("\nクレンジング後のデータを 'cleansed_online_retail.csv' に保存しました。")


except Exception as e:
    print(f"データの読み込みまたはクレンジング中にエラーが発生しました: {e}")
    print("Google Driveの共有設定や列名 (Quantity, UnitPrice, CustomerID) を確認してください。")
```
